## Step 2: Data Preperation
- Dropping irrelevant columns and rows
- Identifying duplicated columns
- Renaming Columns
- Feature Creationion

```python
df.columns
```

```python
# Buat dataframe baru dengan menghapus beberapa kolom
df1 = df[['coaster_name', 'Location', 'Status', 'Manufacturer', \
          'year_introduced', 'latitude', 'longitude', 'Type_Main','opening_date_clean', \
          'speed_mph', 'height_ft', 'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df1.head()
```

```python
df1.info()
```

```python
# ganti tipe data datetime
df1['opening_date_clean'] = pd.to_datetime(df1['opening_date_clean'])
```

```python
# rename kolom
df1 = df1.rename(columns={
    'coaster_name' : 'Coaster_Name',
    'year_introduced' : 'Year_Introduced',
    'latitude' : 'Latitude',
    'longitude' : 'Longitude',
    'opening_date_clean' : 'Opening_Date_Clean',
    'speed_mph' : 'Speed_mph',
    'height_ft' : 'Height_ft',
    'Inversions_clean':'Inversions',
    'Gforce_clean':'Gforce'})
```

```python
df1.isna().sum()
```

```python
df1.duplicated().sum()
```

```python
df1.loc[df1.duplicated(subset=['Coaster_Name'])]
```

```python
# drop koloom duplikat
df1 = df1.drop_duplicates().reset_index(drop=True)
```

## Step 3: Feature Understanding
- Plotting Feature Distribution
- Histogram
- KDE

```python
df1.columns
```

```python
ax = df1['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', 
          title='Top 10 Year Roller Coaster Introduced') 
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Counts')
plt.show()
```

```python
ax = df1['Status'].value_counts().plot(kind='barh', 
                                       title='Number of Status Roller Coaster')
ax.set_xlabel('Counts')
plt.show()
```

```python
ax = df1['Type_Main'].value_counts().plot(kind='bar', 
                                          title='Number of Type Roller Coaster')
ax.set_xlabel('Counts')
plt.show()
```

```python
ax = df1['Speed_mph'].plot(kind='hist', 
                           bins=20, 
                           title='Roller Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.show
```

```python
ax = df1['Speed_mph'].plot(kind='kde', 
                           title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.show()
```

```python
ax = df1['Height_ft'].plot(kind='kde', 
                           title='Coaster Height (ft)')
ax.set_xlabel('Height (ft)')
plt.show()
```

```python
df1.dtypes
```

```python
# num variabel
num_vars = df1.select_dtypes(include=['int','float']).columns.tolist()

# Tentukan tataletak subplot
num_cols = len(num_vars)
num_rows = (num_cols + 2) // 3

# buat subplot
fig, axs = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, 5*num_rows))
axs = axs.flatten()

# plot histogram untuk setiap num variabel
for i, var in enumerate(num_vars):
    df1[var].plot.hist(ax=axs[i])
    axs[i].set_title(var)

if num_cols < len(axs):
    for i in range(num_cols, len(axs)):
        fig.delaxes(axs[i])
        
# sesuaikan layout
fig.tight_layout()

plt.show()
```

## Feature Relationships
- Scatterplot

```python
df1.columns
```

```python
df1.plot(kind='scatter',
         x = 'Speed_mph',
         y = 'Height_ft',
         title='Coaster Speed vs. Height')
plt.show()
```

```python
ax = sns.scatterplot(x='Speed_mph',
                y='Height_ft',
                hue='Year_Introduced',
                data=df1)
ax.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
sns.pairplot(df1,
             vars=['Year_Introduced','Speed_mph',
                   'Height_ft','Inversions','Gforce'])
plt.show()
```

```python
df_corr = df1[['Year_Introduced','Speed_mph',
    'Height_ft','Inversions','Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
plt.title("Correlation between Variables")
plt.show()
```

## Step 5: Ask a Question about the data
- Try to answer a question you have about the data using a plot or statistic.

 
What are the locations with the fastest roller coasters (minimum of 10)?





```python
df1.head()
```

```python
ax = df1.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Avarage Speed (mph)')
plt.show()
```
